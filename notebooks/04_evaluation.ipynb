```python
from choice_model import ChoiceModel
import networkx as nx
model = ChoiceModel()
```

```python
model.graph.nodes[0]
```

# PRODUCER PARTICIPATION INDEX (PPI)

```python
import numpy as np

def cal_PPI():

    def cal_single_PPI(a):
        in_degree = model.graph.in_degree(a)
        out_degree = model.graph.out_degree(a)
        if (in_degree + out_degree) == 0:
            PPI_a = 0
        else:
            PPI_a = out_degree/(in_degree + out_degree)
        return PPI_a

    actors = [n for n in model.graph.nodes() if model.graph.nodes[n]['type'] == 'Actors']
    PPI = np.mean([cal_single_PPI(a) for a in actors])
    return PPI

PPI = cal_PPI()

print(PPI)
```

# COMMUNITY VITALITY INDEX (CVI)

```python
def cal_CVI():
    actors = [n for n in model.graph.nodes() if model.graph.nodes[n]['type'] == 'Actors']
    CVI = np.mean([model.graph.degree(a) for a in actors])
    return CVI

CVI = cal_CVI()
print(CVI)
```

# DECENTRALIZED CONTRIBUTION INDEX (DCI)

```python

def cal_DCI():
    actors = [n for n in model.graph.nodes() if model.graph.nodes[n]['type'] == 'Actors']
    degree_centrality = nx.degree_centrality(model.graph)
    actors_degree_centrality = [degree_centrality[a] for a in actors]
    M = np.mean(actors_degree_centrality)
    IQR = np.percentile(actors_degree_centrality, 75) - np.percentile(actors_degree_centrality, 25)
    DCI = IQR/M
    return DCI

DCI = cal_DCI()
print(DCI)
```
